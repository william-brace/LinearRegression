```python
import numpy as np

x_train = np.array([1,2,3,4,5,6,7])
target = np.array([100,200,100,400,500,800,700])
b = 1

```

So we have the training data and the target/test data. We then have to train the model based on this data. I will be using a linear regression model using mean squared error and gradient descent for this. 

Step 1 is to write down the equation of the line for the regression model. 

We don't have to write the mse directly, only the partical derivative of the cost function which we will use to adjust the weights

Step 2 is to write down the gradient descent alogirthm which we compute the weights with.

```python

w = np.array(1)
y = np.dot(x_train,w) + b

N = len(target)
```

Now we have to train the model using gradient descent. 

```python

lr = 0.01

for i in range(100): 

    y = np.dot(x_train,w) + b
    
    error = y - target
    cost_derivative = np.dot(x_train, error) /N
    
    w = w - lr* cost_derivative
    b = b - lr * np.sum(error) / N

    print(w)
    


print(y)
    
```

Visualizing the model in its current state

```python
import matplotlib.pyplot as plt

plt.scatter(x_train, target, color='blue', label='Actual data')

plt.plot(x_train, y, color='red', label='Model prediction')

# Label the axes
plt.xlabel('Feature')
plt.ylabel('Target')

# Add a title and legend
plt.title('Linear Regression Model')
plt.legend()

# Show the plot
plt.show()


```
